```python
%matplotlib inline
import numpy as np
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt
from pylag import *
```

```python
a = np.arange(0,100,dtype=np.float32)
#x = a
#v = np.exp(-(a-50)**2/2/5**2)*0.1
x = np.concatenate(([0],cumtrapz(1 - np.exp(-(a-49.5)**2/2/5**2)*0.1)))
v = np.zeros((100,),dtype=np.float32)
T = np.ones((100,),dtype=np.float32)

t = 0.
dt = 0.1

for i in range(0,100):
    step(dt,t,x,v,T,f)

plt.figure(figsize = (16,5))
plt.subplot(121)
plt.plot(x, 1/d(x))
plt.subplot(122)
plt.plot(x, v)
```
