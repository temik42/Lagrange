```python
%matplotlib inline
import numpy as np
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt
from pylag import *
```

```python
nx = 64
a = np.arange(0,nx,dtype=np.float32)
#x = a
#v = np.exp(-(a-50)**2/2/5**2)*0.1
x = np.concatenate(([0],cumtrapz(1 - np.exp(-(a-(nx-1)/2.)**2/2/5**2)*0.1)))
#x = a + np.sin(a/(nx-1)*np.pi*2)
v = np.zeros((nx,),dtype=np.float32)
T = np.ones((nx,),dtype=np.float32)

t = 0.
dt = 0.1

for i in range(0,int(1e5)):
    step(dt,t,x,v,T,f)

plt.figure(figsize = (16,5))
plt.subplot(121)
plt.plot(x,1/d(x))
plt.subplot(122)
plt.plot(x,v)

np.sum(T)
```
